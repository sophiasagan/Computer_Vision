```python
import cv2 as cv2
from matplotlib import pyplot as plt 
import numpy as np 
import pandas as pd 
```

Thresholding/Binarization

```python
img = cv2.imread(r'C:\Users\USER\Downloads\Photos\Images\text.jpg')
cv2.imshow('default img', img)
plt.imshow(img)
cv2.waitKey(0)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2.imshow('gray scale images', gray)
plt.imshow(gray)
cv2.waitKey(0)
```

```python
mean_threhsold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 25, 5)
cv2.imshow('thresholding', mean_threhsold)
plt.imshow(cv2.cvtColor(mean_threhsold, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

```python
gauss_threshold = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 25, 5)
cv2.imshow('gauss_thresholding', gauss_threshold)
plt.imshow(cv2.cvtColor(gauss_threshold, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

Morphological Transformation 

```python
img = cv2.imread(r'C:\Users\USER\Downloads\Photos\Images\cats.jpg')
gra = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gauss_threshold = cv2.adaptiveThreshold(gra, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 25, 5)
cv2.imshow('thresholding', gauss_threshold)
plt.imshow(cv2.cvtColor(gauss_threshold, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

```python
img = cv2.imread(r'C:\Users\USER\Downloads\Photos\Images\cats.jpg')
gra = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gauss_blur = cv2.GaussianBlur(gra, (5,5), 3)
cannyMPF = cv2.Canny(gauss_blur, 127, 127)
cannyHPF = cv2.Canny(gauss_blur, 230, 230)
cannyLPF = cv2.Canny(gauss_blur, 21, 21)
cv2.imshow('LPF', cannyLPF)
cv2.imshow('HPF', cannyHPF)
cv2.imshow('MPF', cannyMPF)
plt.imshow(cv2.cvtColor(cannyMPF, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

Laplacian Transformation 

```python
img = cv2.imread(r'C:\Users\USER\Downloads\Photos\Images\park.jpg')
gra = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
canny = cv2.Canny(gra, 127, 127)
lap = cv2.Laplacian(gra, cv2.CV_8U)
lap = np.uint8(np.absolute(lap))
cv2.imshow('gra', lap)
cv2.imshow('canny', canny)
cv2.waitKey(0)
```

SOBEL Derivatives 

```python
SOBEL_X = cv2.Sobel(gra, cv2.CV_64F, 1, 0)
SOBEL_Y = cv2.Sobel(gra, cv2.CV_64F, 0, 1)

cv2.imshow('X-sobel', SOBEL_X)
cv2.imshow('Y_Sobel', SOBEL_Y)

combined_sobel = cv2.bitwise_or(SOBEL_X, SOBEL_Y)
cv2.imshow('combined_sobel',combined_sobel )


cv2.waitKey(0)
```
