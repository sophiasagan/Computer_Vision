Drawing with OpenCV

```python
blank = np.zeros((500, 500, 3), np.uint8)
cv2.imshow('blank', blank)
plt.imshow(blank)
cv2.waitKey(0)
```

```python
print(blank.shape)
```

```python
blank[0:100, 0:500] = (0,0,255)
blank[100:200, 0:500] = (0, 255, 0)
blank[200:300, 0:500] = (255,0, 0)
blank[300:400, 0:500] = (20, 126, 251)
blank[400:500, 0:500] = (230, 145, 187)
cv2.imshow('color_blocks', blank)
cv2.waitKey(0)
```

```python
cv2.rectangle(blank, (0,0), (250, 250), (0,0,0), thickness=1)
plt.imshow(blank)
cv2.imshow('blank', blank)
cv2.waitKey(0)
```

```python
cv2.circle(blank, (blank.shape[1]//2, blank.shape[0]//2), 40, (0,255,0), thickness=2)
plt.imshow(blank)
cv2.imshow('circle', blank)
cv2.waitKey(0)
```

```python
cv2.line(blank, (100, 200), (300, 400), (255,0,0), thickness=2)
plt.imshow(blank)
cv2.imshow('line', blank)
cv2.waitKey(0)
```

```python
cv2.putText(blank, 'Machine Learning India', (0, 200),cv2.FONT_HERSHEY_TRIPLEX, 0.80, (255,255,0), 2)
plt.imshow(blank)
cv2.imshow('Text', blank)
cv2.waitKey(0)
```

Conversion of different color scales 

```python
img = cv2.imread(r'C:\Users\USER\Downloads\Photos\park.jpg')
cv2.imshow('default',img)
cv2.waitKey(0)

vals= cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
cv2.imshow('BGR', vals)
cv2.waitKey(0)

vals = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2.imshow('GRAY', vals)
cv2.waitKey(0)

vals =cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
cv2.imshow('HSV', vals)
cv2.waitKey(0)

vals = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
cv2.imshow('LAB', vals)
cv2.waitKey(0)

vals = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
cv2.imshow('YUV', vals)
cv2.waitKey(0)
```

```python
kernel = np.ones((5,5))/25
print(kernel)
```

```python
print(img.shape)
```

```python
filter2D = cv2.filter2D(img, -1, kernel)
cv2.imshow('filter', filter2D)
plt.imshow(cv2.cvtColor(filter2D, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

```python
plt.hist(filter2D.ravel(), 256, [0,256])
```

```python
avg = cv2.blur(img, (10,5))
cv2.imshow('avg', avg)
plt.imshow(cv2.cvtColor(avg, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

```python
Gaussian_blur = cv2.GaussianBlur(img, (11,11), 16)
cv2.imshow('Gauss', Gaussian_blur)
plt.imshow(cv2.cvtColor(Gaussian_blur, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

```python
median_blur = cv2.medianBlur(img, 11)
cv2.imshow('Gauss', median_blur)
plt.imshow(cv2.cvtColor(median_blur, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```

```python
bilateral_blur = cv2.bilateralFilter(img, 17, 125, 125)
cv2.imshow('bilaty', bilateral_blur)
plt.imshow(cv2.cvtColor(bilateral_blur, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
```
